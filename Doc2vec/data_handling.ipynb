```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
```

```python
template_df = pd.read_csv('./alimtalk_tempalte_parsed.csv', nrows=500000)
template_df.head(4)
```

```python
# 템플릿 상태별 데이터 건수
#Group by
template_df.groupby(by='target_yn').size() #Series 형식으로 리턴
template_df.groupby(by='target_yn').size().reset_index(name='counts') # Dataframe 형태로 리턴
```

```python
#필터링
# Equal
template_df[template_df.target_yn == '정상-발송됨']

# in
template_df[template_df.target_yn.isin(['정상-발송됨', '비정상-검수안됨'])]

# null 필터링
template_df[~template_df.temp_id.isna()]

# 비교연산 필터링
template_df[template_df.temp_id < 10]
template_df[(template_df.temp_id < 10) & (template_df.sender_id < 3)]

#template_df.target_yn in ['정상-발송됨', '비정상-검수안됨']
#template_df.target_yn.isin(['정상-발송됨', '비정상-검수안됨'])
```

```python
#컬럼(변수) 선택
template_df[['temp_id','parsed_content','group_yn','target_yn']]

#컬럼 선택과 필터링 동시에
sample_df = template_df[['sender_id','temp_id','parsed_content','group_yn','target_yn']][template_df.target_yn == '정상-발송됨']

#Sorting
sample_df.sort_values(by='temp_id')#오름차순이 기본
sample_df.sort_values(by='temp_id', ascending=False)
sample_df.sort_values(by=['sender_id','temp_id'], ascending=False)
sample_df.sort_values(by=['sender_id','temp_id'], ascending=False, inplace=True) #Sorting 후 데이터 재저장
```

```python
#기초 통계량 보기
print(sample_df.describe())
print(sample_df.describe(include='all'))
```
